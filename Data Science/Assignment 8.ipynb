```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import *
%matplotlib inline
```

```python
n= 9
d= np.random.poisson(0.7, size=n)
d
```

```python
lamda= np.linspace(0,4,100)    #grid

poster= []

for i in lamda:
  posterior= np.prod(poisson.pmf(d, i)) * uniform.pdf(i,0,4)
  poster.append(posterior)
```
