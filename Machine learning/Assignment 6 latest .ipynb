```python
import numpy as np
import pandas as pd
import scipy.spatial
from collections import Counter
from scipy.spatial import distance
```

```python
df= pd.read_excel('assignment6.xlsx')
df.head()
```

```python
df['versicolor']=pd.get_dummies(df['species'], drop_first=True)
df.head(10)
```

```python
df_test = pd.DataFrame(df,columns=df.columns[:-2])          #train X dataset
X_train= df_test  
X_train.head()
```

```python
y_train= df['versicolor']                         #train y dataset
```

```python
data = {'sepal_length': [8.2], 'sepal_width': [2.6], 'petal_length':[3.2], 'petal_width':[1.8]}  
X_test = pd.DataFrame(data)                         #test X dataset
#X_test.head()
```

```python
class KNN:
    def __init__(self, k):
        self.k = k
        
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        
    
    def predict(self, X_test):
        final_output = []
        for i in range(len(X_test)):
            d = []
            votes = []
            for j in range(len(X_train)):
                dist = scipy.spatial.distance.euclidean(X_train.iloc[[j]] , X_test.iloc[[i]])
                d.append([dist, j])
            d.sort()
            d = d[0:self.k]
            for d, j in d:
                votes.append(y_train[j])
            ans = Counter(votes).most_common(1)[0][0]
            final_output.append(ans)
            
        return final_output
    
    
```

```python
clf = KNN(7)
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)
for i in prediction:
    print(i, end= ' ')
```
